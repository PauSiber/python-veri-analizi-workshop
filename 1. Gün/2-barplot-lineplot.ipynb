```python
import seaborn as sns
import pandas as pd
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
```

```python
df.head()
```

```python
sns.barplot(data = df, x = "cut", y = "price")
```

```python
sns.barplot(data = df, x = "cut", y = "price", hue = "cut", errorbar = None)
```

```python
sns.barplot(data = df, x = "cut", y = "price", hue = "color", errorbar = None)
```
