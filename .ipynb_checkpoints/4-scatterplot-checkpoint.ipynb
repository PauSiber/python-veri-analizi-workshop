```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
```

```python
df.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "day", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "sex", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", size = "size", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", size = "size", hue = "size", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "day",size = "day", data = df);
```
