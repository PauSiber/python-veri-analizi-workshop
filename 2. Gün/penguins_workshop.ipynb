`Adalara gÃ¶re yÃ¼zgeÃ§ uzunluÄŸu daÄŸÄ±lÄ±mÄ±nÄ± Ã§iz.`

```python
sns.boxplot(data=penguins, x="island", y="flipper_length_mm", palette="Set2")
plt.title("Adalara GÃ¶re Flipper UzunluÄŸu DaÄŸÄ±lÄ±mÄ±")
plt.show()
```

`Sadece erkek penguenler iÃ§in aÄŸÄ±rlÄ±k histogramÄ± oluÅŸtur.`

```python
male_penguins = penguins[penguins["sex"] == "Male"]
sns.histplot(data=male_penguins, x="body_mass_g", bins=20, kde=True, color="skyblue")
plt.title("Erkek Penguenlerin AÄŸÄ±rlÄ±k DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("VÃ¼cut AÄŸÄ±rlÄ±ÄŸÄ± (g)")
plt.show()
```

`YÃ¼zgeÃ§ uzunluÄŸu ve gaga uzunluÄŸu arasÄ±ndaki iliÅŸkiyi gÃ¶rselleÅŸtir.`

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", style="sex")
plt.title("Flipper UzunluÄŸu vs Gaga UzunluÄŸu")
plt.show()
```

`Eksik verileri ortalamayla doldurup tekrar scatter plot Ã§iz.`

```python
penguins_filled = penguins.copy()
penguins_filled["bill_length_mm"].fillna(penguins["bill_length_mm"].mean(), inplace=True)
penguins_filled["flipper_length_mm"].fillna(penguins["flipper_length_mm"].mean(), inplace=True)

# Scatter plot
sns.scatterplot(data=penguins_filled, x="flipper_length_mm", y="bill_length_mm", hue="species")
plt.title("Ortalamayla DoldurulmuÅŸ Veride Flipper vs Gaga UzunluÄŸu")
plt.show()
```

Penguenlerin tÃ¼rlerine gÃ¶re ortalama flipper uzunluklarÄ±nÄ± hesaplayÄ±n.

```python
pivot_table = penguins.pivot_table(values="flipper_length_mm", index="species", aggfunc="mean")
print(pivot_table)
```
